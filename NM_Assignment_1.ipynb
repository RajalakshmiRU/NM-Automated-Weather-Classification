Univariate Analysis

```python
sns.histplot(df.Built_Year,kde=True)
```

```python
fig, axarr = plt.subplots(6, figsize=(12,40))
sns.boxplot(y = 'waterfront_present', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[0])
sns.boxplot(y = 'number_of_bedrooms', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[0])
sns.boxplot(y = 'number_of_bathrooms', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[1])
sns.boxplot(y = 'number_of_floors', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[2])
sns.boxplot(y = 'number_of_views', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[3])
sns.boxplot(y = 'condition_of_the_house', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[4])
sns.boxplot(y = 'grade_of_the_house', x = 'Price', data = df,width = 0.8,orient = 'h', showmeans = True, fliersize = 3, ax = axarr[5])
plt.show()
```
